```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ingresos_internet=pd.read_csv('/content/drive/MyDrive/PI_02_intento2/Internet_Ingresos.csv')
velocidad_internet=pd.read_csv('/content/drive/MyDrive/PI_02_intento2/historico_velocidad_internet.csv')
total_nac_acceso=pd.read_csv('/content/drive/MyDrive/PI_02_intento2/Internet_BAF.csv')
ingresos_tel_movil=pd.read_csv('/content/drive/MyDrive/PI_02_intento2/Telefonia_movil.csv')
ingresos_tv=pd.read_csv('/content/drive/MyDrive/PI_02_intento2/Television.csv')
ingresos_tel_fija=pd.read_csv('/content/drive/MyDrive/PI_02_intento2/telefonia_fija.csv')
```

**ESTUDIO DE INGRESOS INTERNET**

```python
ingresos_internet
```

```python
#Mostrar las primeras filas del DataFrame
print(ingresos_internet.head())

#Obtener información sobre el DataFrame
print(ingresos_internet.info())

#Estadísticas resumida de las columnas numéricas
print(ingresos_internet.describe())
```

```python
#Eliminamos puntos por comas y convertimos a tipo float la columna de ingresos
ingresos_internet['Ingresos (miles de pesos)'] = ingresos_internet['Ingresos (miles de pesos)'].str.replace('.', '').astype(float)
```

```python
#renombramos la columna de ingresos
ingresos_internet = ingresos_internet.rename(columns={'Ingresos (miles de pesos)': 'Ingresos_internet (miles de pesos)'})
ingresos_internet
```

```python
# Graficar los ingresos por año utilizando barras verticales con Seaborn
sns.barplot(data=ingresos_internet, x='Año', y='Ingresos_internet (miles de pesos)')

# Agregar etiquetas y título al gráfico
plt.xlabel('Año')
plt.ylabel('Ingresos_internet (miles de pesos)')
plt.title('Ingresos por año de Internet Fijo')

# Mostrar el gráfico
plt.show()
```

Como se observa en el gráfico de Ingresos por año de internet fijo, el mismo ha ido en aumento, lo que significa que hay un crecimiento de los usuarios que adquieren el servicio.

**ESTUDIO DE INGRESOS TELEVISIÓN**

```python
ingresos_tv.columns
```

```python
#Mostrar las primeras filas del DataFrame
print(ingresos_tv.head())

#Obtener información sobre el DataFrame
print(ingresos_tv.info())

#Estadísticas resumida de las columnas numéricas
print(ingresos_tv.describe())
```

```python
#Eliminamos el signo $ en la columna Ingresos TV por suscripción (miles de $) y Ingresos TV satelital (miles de $):
ingresos_tv['Ingresos TV por suscripción  (miles de $)'] = ingresos_tv['Ingresos TV por suscripción  (miles de $)'].str.replace('$', '')
ingresos_tv['Ingresos TV satelital (miles de $)'] = ingresos_tv['Ingresos TV satelital (miles de $)'].str.replace('$', '')
```

```python
#Eliminamos puntos por comas y convertimos a tipo float ambas columnas
ingresos_tv['Ingresos TV por suscripción  (miles de $)'] = ingresos_tv['Ingresos TV por suscripción  (miles de $)'].str.replace('[.,]', '', regex=True).astype(float)
ingresos_tv['Ingresos TV satelital (miles de $)'] = ingresos_tv['Ingresos TV satelital (miles de $)'].str.replace('[.,]', '', regex=True).astype(float)
```

```python
#Graficamos los ingresos por suscripción
sns.lineplot(data=ingresos_tv, x='Año', y='Ingresos TV por suscripción  (miles de $)', label='Ingresos por suscripción')

#Graficamos los ingresos satelitales
sns.lineplot(data=ingresos_tv, x='Año', y='Ingresos TV satelital (miles de $)', label='Ingresos satelitales')

#Agregamos etiquetas y título al gráfico
plt.xlabel('Año')
plt.ylabel('Ingresos (miles de pesos)')
plt.title('Ingresos TV por suscripción y satelitales por Año')

#Mostrar la leyenda
plt.legend()

#Mostramos el gráfico
plt.show()
```

En el gráfico anterior se puede notar que los ingresos de TV por suscripción son superiores a los ingresos por TV satelital

```python
#
import seaborn as sns

# Crear una figura y los subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Graficar ingresos de Internet Fijo
axs[0, 0].bar(x=ingresos_internet['Año'], height=ingresos_internet['Ingresos_internet (miles de pesos)'])
axs[0, 0].set_xlabel('Año')
axs[0, 0].set_ylabel('Ingresos (miles de pesos)')
axs[0, 0].set_title('Ingresos por año de Internet Fijo')

# Graficar ingresos de Telefonía Móvil
axs[0, 1].bar(x=ingresos_tel_movil['Año'], height=ingresos_tel_movil['Ingresos_movil (miles de $)'])
axs[0, 1].set_xlabel('Año')
axs[0, 1].set_ylabel('Ingresos (miles de pesos)')
axs[0, 1].set_title('Ingresos por Año - Telefonía Móvil')

# Graficar ingresos de TV por suscripción y satelitales
axs[1, 0].plot(ingresos_tv['Año'], ingresos_tv['Ingresos TV por suscripción  (miles de $)'], label='Ingresos por suscripción')
axs[1, 0].plot(ingresos_tv['Año'], ingresos_tv['Ingresos TV satelital (miles de $)'], label='Ingresos satelitales')
axs[1, 0].set_xlabel('Año')
axs[1, 0].set_ylabel('Ingresos (miles de pesos)')
axs[1, 0].set_title('Ingresos TV por suscripción y satelitales por Año')
axs[1, 0].legend()

# Graficar ingresos de Telefonía Fija
axs[1, 1].bar(x=ingresos_tel_fija['Año'], height=ingresos_tel_fija['Ingresos_fija (miles de $)'])
axs[1, 1].set_xlabel('Año')
axs[1, 1].set_ylabel('Ingresos (miles de $)')
axs[1, 1].set_title('Ingresos por Año - Telefonía fija')

# Ajustar el espaciado entre subplots
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

Se puede observar, que los ingresos son liderados por internet fijo, seguido por los ingresos de TV por suscripción, luego por los ingresos de telefonía fija, seguido por los ingresos de TV satelital y finalmente los ingresos por telefonía móvil.

**VELOCIDAD DE INTERNET**

```python
velocidad_internet
```

```python
#Renombrar la columna Trimestre.1 a Periodo
velocidad_internet = velocidad_internet.rename(columns={'Trimestre.1': 'Periodo'})

#Mostrar las primeras filas del DataFrame
print(velocidad_internet.head())

#Obtener información sobre el DataFrame
print(velocidad_internet.info())

#Estadísticas resumidas de las columnas numéricas
print(velocidad_internet.describe())
```

```python
#Creamos la figura y un eje
fig, ax = plt.subplots(figsize=(10, 6))

#Graficamos año vs Mbps (Media de bajada)
ax.plot(velocidad_internet['Año'], velocidad_internet['Mbps (Media de bajada)'])

#Invertimos la dirección del eje y
plt.gca().invert_yaxis()

#Agregamos las etiquetas y título al gráfico
ax.set_xlabel('Año')
ax.set_ylabel('Mbps (Media de bajada)')
ax.set_title('Velocidad de Internet - Media de bajada por año')

#Mostramos el gráfico
plt.show()
```

En el gráfico que representa la velocidad del internet a través del tiempo, se observa un aumento de la misma, lo que indica que las empresas mejoran el servicio con el pasar de los años y se mantienen mejorando sus servicios.
